# Relação entre o consumo de energia e os indicadores económicos (US)

Os dados foram extraídos dos sites a seguir:

https://www.census.gov/foreign-trade/statistics/historical/goods.pdf

https://www.census.gov/foreign-trade/statistics/historical/index.html

```python
import pandas as pd
df1 = pd.read_excel(r'goods_usa.xls', index_col=0)
xvalues = list(df1.index)
y1values = df1['Percent Change']
```

Os dados foram extraídos dos sites a seguir:

https://databank.worldbank.org/reports.aspx?source=2&series=NY.GDP.PCAP.KD.ZG#

```python
df2 = pd.read_csv(r'data_mod.csv', index_col=0)
df3 = df2.T
y2values = df3['Manufacturing, value added (annual % growth)']
y3values = df3['Electric power consumption (kWh per capita)']
y4values = df3['Energy use (kg of oil equivalent per capita)']
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fig.subplots_adjust(right=2)

# temos o eixo principal do gráfico (ax): "Balance Change (exports - imports in millions of $)" e os
# outros eixos twin1, twin2 e twin3

twin1 = ax.twinx()
twin2 = ax.twinx()
twin3 = ax.twinx()

twin2.spines["right"].set_position(("axes", 1.1))
twin3.spines["right"].set_position(("axes", 1.2))

# eixo principal
p1, = ax.plot(xvalues, y1values, "k-", label="Balance Change (exports - imports in millions of $)")

# eixos secundários
p2, = twin1.plot(xvalues, y2values, "tab:gray", label="Manufacturing, value added (annual % growth)")
p3, = twin2.plot(xvalues, y3values, "b-", label="Electric power consumption (kWh per capita)")
p4, = twin3.plot(xvalues, y4values, "r-", label="Energy use (kg of oil equivalent per capita)")

ax.set_xlim(1998, 2014)
ax.set_ylim(-40, 40)
twin1.set_ylim(-10, 10)
twin2.set_ylim(12900, 13800)
twin3.set_ylim(6800, 8100)

# aquí colocamos os nomes de cada eixo
ax.set_xlabel("")
ax.set_ylabel("Balance Change (exports - imports in millions of $)")
twin1.set_ylabel("Manufacturing, value added (annual % growth)")
twin2.set_ylabel("Electric power consumption (kWh per capita)")
twin3.set_ylabel("Energy use (kg of oil equivalent per capita)")

ax.yaxis.label.set_color(p1.get_color())
twin1.yaxis.label.set_color(p2.get_color())
twin2.yaxis.label.set_color(p3.get_color())
twin3.yaxis.label.set_color(p4.get_color())

tkw = dict(size=4, width=1)
ax.tick_params(axis='y', colors=p1.get_color(), **tkw)
twin1.tick_params(axis='y', colors=p2.get_color(), **tkw)
twin2.tick_params(axis='y', colors=p3.get_color(), **tkw)
twin3.tick_params(axis='y', colors=p4.get_color(), **tkw)
ax.tick_params(axis='x', **tkw)

#ax.legend(handles=[p1, p2, p3])
plt.xticks([1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014])
plt.title('U.S. energy consumption and economic indicators correlation by year')
#ax.tick_params(axis='x', labelsize=7)

plt.show()
```

Agora, no gráfico de 4 eixos y, podemos observar como os indicadores econômicos da economia dos EEUU (Balance change e value added) se relacionam com o consumo de energia (elétrica e hidrocarburos) no período 1998 - 2014.
